```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn import linear_model
%matplotlib inline
```

```python
data = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Credit.csv', index_col=0)
```

```python
X=data[['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education']]
Y=data['Balance']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=1)
```

```python
t=np.logspace(1,6)
```

```python
coe=[]
inter=[]
rcuadrado=[]
for i in t:
    clf = linear_model.Lasso(alpha=i)
    clf.fit(X_train, Y_train)
    coe.append(clf.coef_)
    inter.append(clf.intercept_)
    rcuadrado.append(clf.score(X_test, Y_test))
coe=np.array(coe)
inter=np.array(inter)
rcuadrado=np.array(rcuadrado)
```

```python
plt.semilogx(t,rcuadrado)
```

```python
plt.figure()
plt.semilogx(t,coe[:,0],label='beta 1')
plt.semilogx(t,coe[:,1],label='beta 2')
plt.semilogx(t,coe[:,2],label='beta 3')
plt.semilogx(t,coe[:,3],label='beta 4')
plt.semilogx(t,coe[:,4],label='beta 5')
plt.semilogx(t,coe[:,5],label='beta 6')
plt.legend()
```

```python
plt.figure()
plt.semilogx(t,inter,label='beta 0')
```

```python
scaler = preprocessing.StandardScaler().fit(X_test)
scaler1 = preprocessing.StandardScaler().fit(X_train)

Y_test= (Y_test-np.mean(Y_test)) / np.std(Y_test)
Y_train= (Y_train-np.mean(Y_train)) / np.std(Y_train)
```

```python
X_train=scaler1.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
t=np.linspace(0.0001,1)
coeNor=[]
interNor=[]
rcuadradoNor=[]
for i in t:
    clf = linear_model.Lasso(alpha=i)
    clf.fit(X_train, Y_train)
    coeNor.append(clf.coef_)
    interNor.append(clf.intercept_)
    rcuadradoNor.append(clf.score(X_test, Y_test))
coeNor=np.array(coeNor)
interNor=np.array(interNor)
rcuadradoNor=np.array(rcuadradoNor)
```

```python
plt.figure()
plt.semilogx(t,rcuadradoNor)
```

```python
plt.figure()
plt.semilogx(t,coeNor[:,0],label='beta 1')
plt.semilogx(t,coeNor[:,1],label='beta 2')
plt.semilogx(t,coeNor[:,2],label='beta 3')
plt.semilogx(t,coeNor[:,3],label='beta 4')
plt.semilogx(t,coeNor[:,4],label='beta 5')
plt.semilogx(t,coeNor[:,5],label='beta 6')
plt.legend()
```

```python
plt.figure()
plt.semilogx(t,interNor,label='beta 0')
```
